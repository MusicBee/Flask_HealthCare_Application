```python
### IMPORT RELATED LIBRARIES TO PERFORM TASKS
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import os
```

```python
### LOAD USER_DATA.CSV INTO DATA FRAME

df = pd.read_csv("/content/sample_data/user_data.csv")
df.head()
```

```python
### SHOW AGES WITH HIGHEST INCOME

# Sort by total_income descending
df_sorted = df.sort_values(by="total_income", ascending=False)

# Plot ages vs income
plt.figure(figsize=(10, 6))
plt.bar(df_sorted['age'].astype(str), df_sorted['total_income'])
plt.title("Ages with the Highest Income", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Total Income (₦)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display the chart
plt.show()
plt.savefig("ages_highest_income.png", bbox_inches='tight')
```

```python
#### GENDER DISTRIBUTION ACCROSS SPENDING CATEGORIES

# Group by gender and sum each expense category
expense_categories = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
gender_expenses = df.groupby('gender')[expense_categories].sum()

# Plot grouped bar chart
gender_expenses.T.plot(kind='bar', figsize=(10, 6))
plt.title("Gender Distribution Across Spending Categories", fontsize=14)
plt.xlabel("Expense Categories")
plt.ylabel("Total Amount Spent (₦)")
plt.xticks(rotation=30)
plt.legend(title="Gender")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show and Save the chart
plt.show()
plt.savefig("gender_distribution_accross_cat.png", bbox_inches='tight')
```
